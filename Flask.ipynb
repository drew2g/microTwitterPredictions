```python
#global twitter feed
Gnew_tweets = ""
GsearchQuery = ""
Gtrends = ""
```

```python
#refresh method that modifies global variables
def getData():
    trends1 = api.trends_place(1)
    data = trends1[0] 
    trends = data['trends']
    searchQuery = trends[0]['name']
    new_tweets = api.search(q=searchQuery,
                            result_type = "recent",
                            lang = "en") 
    global Gnew_tweets
    global GsearchQuery
    global Gtrends
    
    Gnew_tweets = new_tweets
    GsearchQuery = searchQuery
    Gtrends = trends
    
    return new_tweets, searchQuery, trends
```

```python
#background thread refreshing api
import threading
import time

class APIReload(object):
    def __init__(self, interval=5): #interval in seconds
        self.interval = interval
        thread = threading.Thread(target=self.run, args=())
        thread.daemon = True                            
        thread.start()                                  

    def run(self):
        while True:
            getData()
            time.sleep(self.interval)
            
APIReload()
```
